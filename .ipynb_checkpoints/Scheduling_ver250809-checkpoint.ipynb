```python
from gurobipy import *
import numpy as np
import pandas as pd
import math
import time
from datetime import datetime, timedelta, date
import holidays
from calendar import isleap
```

#### Creating Model

```python
m = Model('clothing manufacturing model')
```

#### Setting start date

```python
# get days in each month
def get_day_dict(year):
    # Base dictionary for days in each month
    day_dict = {1: 31, 2: 28, 3: 31, 4: 30, 5: 31, 6: 30,
                7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31}
    # Adjust February for leap years
    if isleap(year):
        day_dict[2] = 29
    return day_dict
```

```python
year = 2024
month = 8
day = 1
start_date = datetime(year, month, day) 
```

```python
# holidays in Taiwan
tw_holidays = holidays.TW(years=year)
print(tw_holidays)
```

#### Setting Order Number

```python
# 8, 10, 12, 14, 16
order_num = 16
```

#### Creating Set

```python
day_dict = get_day_dict(year)
Num_L = day_dict[month]

# set of holidays and weekends
HL = set()
for day in range(1, day_dict[month] + 1):
    current_date = date(year, month, day)
    if current_date.weekday() >= 5 or current_date in tw_holidays:  # Check for weekends or holidays
        HL.add(day - 1) # day index is start from zero
print("holiday:", HL)

# dict of day no. and date
L_date = dict()
no = 0 
for l in range(Num_L):
    if l not in HL:
        L_date[no] = l
        no += 1

print(L_date)
L = set(L_date.keys())
print(L)

Num_K = order_num
K = set(range(Num_K))
Num_D = 2
D = set(range(Num_D))
print(K)

# the number of equipments 
Num_Ed = [10, 10]
# the number of all equipments
Num_E = sum(Num_Ed)
# equipment set
Ed = dict()
i = 0
for d in D:
    Ed[d] = set()
    while len(Ed[d]) < Num_Ed[d]:
        Ed[d].add(i)
        i = i + 1

print(Ed)
# all equipment set
E = set(range(Num_E))

## notice bearing index
# the number of bearings
Num_Pd = [2, 3]
# the number of all bearings
Num_P = sum(Num_Pd)
# bearing set
Pd = dict()
i = 0
for d in D:
    Pd[d] = set()
    while len(Pd[d]) < Num_Pd[d]:
        Pd[d].add(i)
        i = i + 1
        
print(Pd)
# all bearing set
P = set(range(Num_P))

Num_I = 2
I = set(range(Num_I))
print(I)
Num_J = 3
J = set(range(Num_J))
print(J)

# LP-660: 4
# LP-770: 5
```

#### Creating Parameter

```python
# Reading Order Data
# order quantity
with open(f"./data/order/{order_num}/quantity.txt", 'r') as file:
    # Read the content of the file and split it 
    content = file.read()
    q = list(map(int, content.split()))

# order value
with open(f"./data/order/{order_num}/value.txt", 'r') as file:
    # Read the content of the file and split it 
    content = file.read()
    v = list(map(int, content.split()))

# order due date
with open(f"./data/order/{order_num}/due_date.txt", 'r') as file:
    # Read the content of the file and split it 
    content = file.read()
    u = list(map(int, content.split()))

print(q)
print(v)
print(u)
```

```python
# q = [800, 800, 800, 800, 800, 800, 800, 800]
# v = [111942.13, 140985.87, 138375.47, 149479.2, 145684, 151972, 109269.33, 106880.8]
# u = [22, 27, 17, 15, 30, 36, 17, 20, 13, 10, 12] 
r = 0.01
a = [3, 5]

b = 100 
M = 9999                                
```

#### Adjust due date (to align with zero-started day index)

```python
for i, dd in enumerate(u):
    new_dd = dd - 1 # day index starts from zero 
    u[i] = new_dd
```

```python
print (u)
```

#### Creating Variable

```python
x = m.addVars(Num_L, Num_K, Num_I, Num_E, Num_P, Num_J - 1, vtype=GRB.BINARY, name="x")
y = m.addVars(Num_L, Num_K, Num_I, Num_E, Num_P, Num_J, vtype=GRB.BINARY, name="y")
t = m.addVars(Num_K, Num_I, vtype=GRB.INTEGER, name="t")
w = m.addVars(Num_K, Num_I, vtype=GRB.INTEGER, name="w")
c = m.addVars(Num_K, vtype=GRB.INTEGER, name="c")
```

#### Update

```python
m.update()
```

#### Objective Function

```python
# summation vk
sum_v = LinExpr()
for k in K:
    sum_v += v[k]

# summation vk * r * ck
sum_vrc = LinExpr()
for k in K:
    sum_vrc += v[k] * r * c[k]

sum_ck = LinExpr()
for k in K:
    sum_ck += c[k]
    
# setting obj.
# m.setObjective(sum_v - sum_vrc, GRB.MAXIMIZE) 
m.setObjective(sum_vrc, GRB.MINIMIZE)
```

#### Constraints

```python
# constraint 1
for l in L:
    for d in D:
        for e in Ed[d]:
            for p in Pd[d]:
                c1_sum_x = LinExpr()
                for k in K:
                    for i in I:
                        # excluding j = 2
                        for j in J:
                            if j != 2:
                               c1_sum_x += x[l, k, i, e, p, j]

                m.addConstr(c1_sum_x <= 1, "c1")
```

```python
# constraint 2
for l in L:
    for d in D:
        for e in Ed[d]:
            for p in Pd[d]:
                c2_sum_y = LinExpr()
                for k in K:
                    for i in I:
                        for j in J:
                           c2_sum_y += y[l, k, i, e, p, j]

                m.addConstr(c2_sum_y <= 1, "c2")
```

```python
# constraint 3
for k in K:
    for i in I:
        for j in J:
            if j != 2:
                c3_sum_x = LinExpr()
                for l in L:
                    for d in D:
                        for e in Ed[d]:
                            for p in Pd[d]:
                                c3_sum_x += x[l, k, i, e, p, j]
    
                m.addConstr(c3_sum_x == 1, "c3")
```

```python
# constraint 4
for k in K:
    for i in I:
        for d in D:
            for e in Ed[d]:
                for p in Pd[d]:
                    c4_sum_x0 = LinExpr()
                    c4_sum_x1 = LinExpr()
                    for l in L:
                        c4_sum_x0 += x[l, k, i, e, p, 0]
                        c4_sum_x1 += x[l, k, i, e, p, 1]

                    m.addConstr(M * c4_sum_x0 >= c4_sum_x1, "c4")
```

```python
# constraint 5
for k in K:
    for i in I:
        for d in D:
            for e in Ed[d]:
                for p in Pd[d]:
                    c5_sum_x0 = LinExpr()
                    c5_sum_x1 = LinExpr()
                    for l in L:
                        c5_sum_x0 += (l + a[0]) * x[l, k, i, e, p, 0]
                        c5_sum_x1 += l * x[l, k, i, e, p, 1]

                    m.addConstr(c5_sum_x1 == c5_sum_x0, "c5")
```

```python
# constraint 6
for k in K:
    for i in I:
        for d in D:
            for e in Ed[d]:
                for p in Pd[d]:
                    for j in J:
                        if j != 2:
                            for l in range(Num_L - a[j] + 1):
                                c6_sum_y = LinExpr()
                                for l_p in range(l, l + a[j]):
                                    c6_sum_y += y[l_p, k, i, e, p, j]

                                m.addConstr(c6_sum_y >= - M * (1 - x[l, k, i, e, p, j]) + a[j], "c6_r")
                                m.addConstr(c6_sum_y <= M * (1 - x[l, k, i, e, p, j]) + a[j], "c6_l")
                                                            
```

```python
# constraint 7
for k in K:
    for i in I:
        for d in D:
            for e in Ed[d]:
                for p in Pd[d]:
                    for j in J:
                        if j != 2:
                            c7_sum_x = LinExpr()
                            c7_sum_y = LinExpr()
                            for l in L:
                                c7_sum_x += x[l, k, i, e, p, j]
                                c7_sum_y += y[l, k, i, e, p, j]

                            m.addConstr(c7_sum_x * a[j] >= c7_sum_y, "c7")
                                                     
```

```python
# constraint 8
for k in K:
    for i in I:
        for l_p in L:
            for d_p in D:
                for e_p in Ed[d_p]:
                    for p_p in Pd[d_p]:
                        c8_sum_x = LinExpr()
                        for l in L:
                            for d in D:
                                for e in Ed[d]:
                                    for p in Pd[d]:
                                        c8_sum_x += x[l, k, i, e, p, 1] * (l + a[1])

                        m.addConstr(
                            M * (1 -  y[l_p, k, i, e_p, p_p, 2]) + l_p * y[l_p, k, i, e_p, p_p, 2] >= c8_sum_x, "c8")
                                       
```

```python
# constraint 9
for k in K:
    for i in I:
        c9_sum_y = LinExpr()
        for l in L:
            for d in D:
                for e in Ed[d]:
                    for p in Pd[d]:
                        c9_sum_y += y[l, k, i, e, p, 2]

        m.addConstr(c9_sum_y == math.ceil(q[k] / b), "c9")
                                       
                    
```

```python
# constraint 10
for l in L:
    c10_sum_y = LinExpr()
    for k in K:
        for i in I:
            for j in J:
                for d in D:
                    for e in Ed[d]:
                        for p in Pd[d]:
                            c10_sum_y += y[l, k, i, e, p, j]

    m.addConstr(c10_sum_y <= 50, "c10")
```

```python
# constraint 11
for l in L:
    for k in K:
        for i in I:
            for d in D:
                for e in Ed[d]:
                    for p in Pd[d]:
                        # using date
                        m.addConstr(L_date[l] * y[l, k, i, e, p, 2] - u[k] <= t[k, i], "c11")
```

```python
# constraint 12 
for k in K:
    for i in I:
        # gurobi doesn't support strict inequality, so we add 1e-5 in first constraint
        m.addConstr(7 * (w[k, i] - 1) + 1e-5 <=  t[k, i], "c12_l")
        m.addConstr(7 * w[k, i] >= t[k, i], "c12_r")
```

```python
# constraint 13
for k in K:
    for i in I:
        m.addConstr(w[k, i] <= c[k], "c13")
```

#### Result

```python
# record cpu time
start_time = time.time()

# set time limit 600 sec
m.setParam('TimeLimit', 1800)
m.optimize()

# End the timer 
cpu_time = time.time() - start_time
```

#### Generating Excel

```python
# Create dataframe 
tuple_list = []
for d in D:
    for e in Ed[d]:
        for p in Pd[d]:
            if d > 0:
                bearing = p - sum(Num_Pd[:d])
            else:
                bearing = p
            tuple_list.append((f"Equipment {e}", f"Bearing {bearing}" ))

col_name = pd.MultiIndex.from_tuples(tuple_list, names=['Equipment', 'Bearing'])
# using all date
df = pd.DataFrame(index=[f"Day {day}" for day in range(Num_L)], columns = col_name)
```

```python
df
```

```python
# Fill Scheduling DataFrame
for k in K:
    for i in I:
        for j in J:
            for d in D:
                for e in Ed[d]:
                    for p in Pd[d]:       
                        for l in L:
                            if y[l, k, i, e, p, j].x > 0:
                                if i == 0: 
                                    side = "A"
                                else:
                                    side = "B"

                                if d > 0:
                                    bearing = p - sum(Num_Pd[:d])
                                else:
                                    bearing = p
                                     
                                df.at[f"Day {L_date[l]}", (f"Equipment {e}", f"Bearing {bearing}") ] = f"Order {k} Side {side} Stage {j + 1}"
                                # print(f"Order {k} Side {i} stage {j} is printing on Equipment {e} Bearing {p} at time {l}")
```

```python
# Replace day no. with date 
df.index = df.index.map(lambda x: (start_date + timedelta(days=int(x.split()[1]))).strftime('%m/%d'))
```

```python
# Use color to represent different orders
color_pairs = {
    'Order 0': ('#1f77b4', '#aec7e8'),
    'Order 1': ('#ff7f0e', '#ffbb78'),
    'Order 2': ('#2ca02c', '#98df8a'),
    'Order 3': ('#d62728', '#ff9896'),
    'Order 4': ('#9467bd', '#c5b0d5'),
    'Order 5': ('#8c564b', '#c49c94'),
    'Order 6': ('#e377c2', '#f7b6d2'),
    'Order 7': ('#7f7f7f', '#c7c7c7'),
    'Order 8': ('#bcbd22', '#dbdb8d'),
    'Order 9': ('#17becf', '#9edae5'),
    'Order 10': ('#393b79', '#5254a3'),
    'Order 11': ('#637939', '#8ca252'),
    'Order 12': ('#8c6d31', '#bd9e39'),
    'Order 13': ('#843c39', '#ad494a'),
    'Order 14': ('#7b4173', '#a55194'),
    'Order 15': ('#17becf', '#98df8a'),
}
```

```python
# Save the Excel
with pd.ExcelWriter(f"./schedule_result/result/{order_num}/scheduling_gurobi_order_{order_num}_practical.xlsx") as writer:
    df.to_excel(writer, sheet_name="Schedule")
    
    # Access the xlsxwriter workbook and worksheet objects
    workbook  = writer.book
    worksheet = writer.sheets["Schedule"]
    
    # Set the width of each column
    for col_num, col in enumerate(df.columns, start=1):
        worksheet.set_column(col_num, col_num, 25)

    # set color for each order 
    for col_num, col in enumerate(df.columns, start=1):
        for row_num, value in enumerate(df[col], start=3):
            if pd.notna(value):
                # split the value, which is Order k Side i Stage j
                split_str = value.split()
                order_key = split_str[0] + " " + split_str[1]

                if "A" in split_str :
                    color = color_pairs[order_key][0]
                else:
                    color = color_pairs[order_key][1]
                          
                worksheet.write(row_num, col_num, value, workbook.add_format({'bg_color': color}))
```
